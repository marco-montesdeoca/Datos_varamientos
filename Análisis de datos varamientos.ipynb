## Análisis de datos varamientos

```python
# Usamos la librería pandas para leer archivos csv y excel en python
import pandas as pd
# Cargamos los archivos en el objeto "df_in"
df_in = pd.read_excel('biodiversidad-4.xlsx') # archivo convertido a formato excel antes de abrir aquí
```

```python
# Revisamos las columnas totales
for name in df_in.columns.values:
    print('* ', name)
```

De todas las columnas, por ahora continuaremos solo con las siguientes: <br>

* Región
* Cantidad de Ejemplares
* Tipo de Especie

```python
# Extraemos columnas de interés en un nuevo dataframe "df2"
df2 = df_in[['Región', 'Cantidad de Ejemplares', 'Tipo de Especie']]
```

```python
# Revisamos la primera fila
df2.head(1)
```

```python
# Convertimos el dataframe en una tabla tipo spreadsheet, donde
# las especies son columnas y las regiones son filas

tabla3 = pd.pivot_table(df2, values='Cantidad de Ejemplares', 
                        index='Región', columns='Tipo de Especie', aggfunc='sum')
```

```python
# Últimos ajustes: reemplazar valores ausentes por ceros
tabla3.fillna(0,inplace=True)
```

```python
# Últimos ajustes: reemplazar valores en formato decimal a formato número entero
tabla3 = tabla3.astype(int)
```

#### Resultado final
Table 3. Number of stranded specimens by species and geographical region

```python
tabla3
```

Al revisar la tabla resumen, podemos ver que la especie más afectada corresponde a **PI - Lobo Marino Común, Lobo de Un Pelo - Otaria flavescens**.

```python
# Exportamos a archivo excel
tabla3.to_excel("Tabla resumen varamientos.xlsx")
```
